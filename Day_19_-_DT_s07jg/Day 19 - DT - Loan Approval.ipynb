```python
# The user will pass the credit score and the model will confirm if the loan needs to be approved or not
```

```python
import pandas as pd
credit_score = [650, 720, 580, 800, 690, 750, 600, 670, 710, 680]
loan_approval = [1, 1, 0, 1, 0, 1, 0, 1, 1, 0]

data = {"Credit_score" : credit_score, "Loan_Approval" : loan_approval}
df = pd.DataFrame(data)
df
```

```python
X = df[['Credit_score']]
y = df[['Loan_Approval']]
```

```python
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X , y)
```

```python
user_credit_score = int(input("Enter your credit score:"))
pred = clf.predict([[user_credit_score]])
if (pred[0] == 1):
    print("Congratulations ! Your loan is approved.")
else:
    print("Your application is rejected.")
```

```python
pred[0]
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plot_tree(clf, feature_names=['Credit score'], class_names=['Denied', 'Approved'], filled=True)
plt.show()
```
